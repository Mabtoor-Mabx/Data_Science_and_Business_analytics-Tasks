# 5-Analyze The data

### I-KDE Plot 

```python
plt.figure(figsize=(15,3))
sns.kdeplot(super_store_market['Sales'], color='Red', label='Sales',shade=True, bw_method=25)
sns.kdeplot(super_store_market['Profit'], color='Black', label='Profit', shade=True, bw_method=25)
plt.xlim([-100,1000])
plt.legend()
plt.show()
```

### II-Pairplot  (Based on Category)

```python
sns.pairplot(super_store_market,hue='Category')
plt.show()
```

### III- Pairpolt (Based on Region)

```python
sns.pairplot(super_store_market,hue='Region')
plt.show()
```

### IV-Correlation

```python
super_store_market.corr()
```

### V- Heatmap of Correlation 

```python
sns.heatmap(super_store_market.corr(), annot=True)
plt.show()
```

**As you can see in Heat map:**

**1- Sales and Profit are moderately Correlated**

**2-Quantity and Profit  are moderately correlated**

**3-Discount and Profit are negative correlated**

### VI- Countplot of columns 

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,7));

sns.countplot(super_store_market['Category'],ax=axs[0][0])
sns.countplot(super_store_market['Segment'],ax=axs[0][1])
sns.countplot(super_store_market['Ship Mode'],ax=axs[1][0])
sns.countplot(super_store_market['Region'],ax=axs[1][1])
axs[0][0].set_title('Category',fontsize=20)
axs[0][1].set_title('Segment',fontsize=20)
axs[1][0].set_title('Ship Mode',fontsize=20)
axs[1][1].set_title('Region',fontsize=20)

plt.tight_layout()
plt.show()
```

### VII- Countplot of Subcategory 

```python
plt.figure(figsize=(18,6))
sns.countplot(super_store_market['Sub-Category'])
plt.title('Sub-Category',fontsize=20)
plt.show()
```

### VIII- Countplot of Quantity 

```python
plt.figure(figsize=(18,6))
sns.countplot(super_store_market['Quantity'])
plt.title('Quantity',fontsize=20)
plt.show()
```

### IX- Countplot of Discount

```python
plt.figure(figsize=(18,6))
sns.countplot(super_store_market['Discount'])
plt.xticks(rotation=90)
plt.title('Discount',fontsize=20)
plt.show()
```

### X-  Distribution of data using plot

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (15,12))
sns.distplot(super_store_market['Sales'], color = 'red',  ax = axs[0][0])
sns.distplot(super_store_market['Profit'], color = 'green',  ax = axs[0][1])
sns.distplot(super_store_market['Quantity'], color = 'orange',  ax = axs[1][0])
sns.distplot(super_store_market['Discount'], color = 'blue',  ax = axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 20)
axs[0][1].set_title('Profit Distribution', fontsize = 20)
axs[1][0].set_title('Quantity distribution', fontsize = 20)
axs[1][1].set_title('Discount Distribution', fontsize = 20)
plt.show()
```
