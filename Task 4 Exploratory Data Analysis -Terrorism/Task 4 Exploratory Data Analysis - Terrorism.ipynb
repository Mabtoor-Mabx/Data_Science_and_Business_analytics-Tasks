# 4- Check if there is some null values or not? 

```python
global_terrorism_dataset.isnull().sum()
```

**As we can see, There is some Missing values in our dataset, we have to remove these missing values to clean our data**

```python
global_terrorism_dataset['Wound'] = global_terrorism_dataset['Wound'].fillna(0)
global_terrorism_dataset['kill'] = global_terrorism_dataset['kill'].fillna(0)
```

# 5- Data Visualization

### I- BarPlot

```python
year = global_terrorism_dataset['Year'].unique()
years_count = global_terrorism_dataset['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

### II- CrossTab

```python
pd.crosstab(global_terrorism_dataset.Year, global_terrorism_dataset.Region).plot(kind='area',stacked=False,figsize=(20,5))
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
attack = global_terrorism_dataset.Country.value_counts()[:10]
attack
```

```python
global_terrorism_dataset.Group.value_counts()[1:10]
```

### III- Country Wise BarPlot 

```python
plt.subplots(figsize=(20,10))
sns.barplot(global_terrorism_dataset['Country'].value_counts()[:10].index,global_terrorism_dataset['Country'].value_counts()[:10].values,palette='YlOrBr_r')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

### IV-PeopleWise Barplot 

```python
global_terrorism_dataset_attack = global_terrorism_dataset[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
global_terrorism_dataset_attack.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.title("People Died Due To Attack",fontsize=25)
plt.ylabel("Number of killed peope",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

### V- Most Effected Countries BarPlot 

```python
global_terrorism_dataset['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected city",fontsize=20)
plt.show()
```

### VI- Killed Person BarPlot 

```python
global_terrorism_dataset[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,5),color=['red'])
plt.xticks(rotation=50)
plt.title("Number of killed ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

### VII- Wounded People BarPlot 

```python
global_terrorism_dataset[['Attacktype','Wound']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,5),color=['green'])
plt.xticks(rotation=50)
plt.title("Number of wounded  ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

### VIII- Per Year Attack BarPlot 

```python
plt.subplots(figsize=(20,5))
sns.countplot(global_terrorism_dataset["Targettype"],order=global_terrorism_dataset['Targettype'].value_counts().index,palette="gist_heat",edgecolor=sns.color_palette("mako"));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

### IX- Terrorist Group Attack Number BarPlot 

```python
global_terrorism_dataset['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='darkorange',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("Attack number",fontsize=15)
plt.show()
```

### X- Terrorist  Attack Killed People BarPlot 

```python
global_terrorism_dataset[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',color='black',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```
